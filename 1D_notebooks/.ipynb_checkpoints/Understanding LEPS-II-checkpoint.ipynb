```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
cap = 700
def Q_orig(d, alpha, r, r0):
    exponent = -2 * alpha * (r - r0)
    if exponent > cap:
        return (d / 2) * (1.5 * np.exp(cap) - np.exp(cap // 2))
    return (d / 2) * (1.5 * np.exp(-2 * alpha * (r - r0)) - np.exp(-alpha * (r - r0)))
    
def J_orig(d, alpha, r, r0):
    exponent = -2 * alpha * (r - r0)
    cap_local = 350
    if exponent > cap_local:
        return (d / 4) * (1.5 * np.exp(cap_local) - 6 * np.exp(cap_local // 2))
    return (d / 4) * (np.exp(-2 * alpha * (r - r0)) - 6 * np.exp(-alpha * (r - r0)))

def V_LEPS(rAB, rBC, debug = False):
    a = 0.05
    b = 0.30
    c = 0.05
    dAB = dBC = 4.746
    dAC = 3.445
    r0 = 0.742
    alpha = 1.942
    QAB = Q_orig(dAB, alpha, rAB, r0)
    QBC = Q_orig(dBC, alpha, rBC, r0)
    rAC = rAB + rBC
    QAC = Q_orig(dAC, alpha, rAC, r0)
    Q_values = (QAB / (1 + a)) + (QBC / (1 + b)) + (QAC / (1 + c))
    if debug:
        print("QAB / (1 + a) : ", (QAB / (1 + a)))
        print("QBC / (1 + b) : ", (QBC / (1 + b)))
        print("QAC / (1 + c) : ", (QAC / (1 + c)))
    JAB = J_orig(dAB, alpha, rAB, r0)
    JBC = J_orig(dBC, alpha, rBC, r0)
    JAC = J_orig(dAC, alpha, rAC, r0)
    if debug:
        print("JAB**2 / (1 + a)**2 : ", JAB**2 / (1 + a)**2)
        print("JBC**2 / (1 + b)**2 : ", JBC**2 / (1 + b)**2)
        print("JAC**2 / (1 + c)**2 : ", JAC**2 / (1 + c)**2)
        print("JAB * JBC / (1 + a) * (1 + b) : ", (JAB*JBC/((1+a)*(1+b))))
        print("JBC * JAC / (1 + b) * (1 + c) : ", (JBC*JAC/((1+b)*(1+c))))
        print("JAB * JAC / (1 + a) * (1 + c) : ", (JAB*JAC/((1+a)*(1+c))))
    
    J_values = (JAB / (1 + a))**2 + (JBC / (1 + b))**2 + (JAC / (1 + c))**2
    
    
    J_values = J_values - ((JAB*JBC/((1+a)*(1+b))) + (JBC*JAC/((1+b)*(1+c))) + (JAB*JAC/((1+a)*(1+c))))
    if debug:
        print("sqrt(J_values) : ", np.sqrt(J_values))
    return Q_values - np.sqrt(J_values)

def V_LEPS_II(rAB, x, debug = False):
    rAC = 3.742
    kC = 0.2025
    V_normal = V_LEPS(rAB, rAC - rAB, debug)
    c = 1.154
    if debug:
        print("Harmonic : ", 2 * kC * (rAB - (rAC / 2 - x / c))**2)
    return V_normal + 2 * kC * (rAB - (rAC / 2 - x / c))**2
```

### Modifying the LEPS_II potential to increase barrier height

```python
def V_LEPS_mod(rAB, rBC, debug = False):
    a = 0.05
    b = 0.03
    c = 0.05
    dAB = dBC = 4.746
    dAC = 3.445
    r0 = 0.742
    alpha = 1.942
    QAB = Q_orig(dAB, alpha, rAB, r0)
    QBC = Q_orig(dBC, alpha, rBC, r0)
    rAC = rAB + rBC
    QAC = Q_orig(dAC, alpha, rAC, r0)
    Q_values = (QAB / (1 + a)) + (QBC / (1 + b)) + (QAC / (1 + c))
    if debug:
        print("QAB / (1 + a) : ", (QAB / (1 + a)))
        print("QBC / (1 + b) : ", (QBC / (1 + b)))
        print("QAC / (1 + c) : ", (QAC / (1 + c)))
    JAB = J_orig(dAB, alpha, rAB, r0)
    JBC = J_orig(dBC, alpha, rBC, r0)
    JAC = J_orig(dAC, alpha, rAC, r0)
    if debug:
        print("JAB**2 / (1 + a)**2 : ", JAB**2 / (1 + a)**2)
        print("JBC**2 / (1 + b)**2 : ", JBC**2 / (1 + b)**2)
        print("JAC**2 / (1 + c)**2 : ", JAC**2 / (1 + c)**2)
        print("JAB * JBC / (1 + a) * (1 + b) : ", (JAB*JBC/((1+a)*(1+b))))
        print("JBC * JAC / (1 + b) * (1 + c) : ", (JBC*JAC/((1+b)*(1+c))))
        print("JAB * JAC / (1 + a) * (1 + c) : ", (JAB*JAC/((1+a)*(1+c))))
    
    J_values = (JAB / (1 + a))**2 + (JBC / (1 + b))**2 + (JAC / (1 + c))**2
    
    
    J_values = J_values - ((JAB*JBC/((1+a)*(1+b))) + (JBC*JAC/((1+b)*(1+c))) + (JAB*JAC/((1+a)*(1+c))))
    if debug:
        print("sqrt(J_values) : ", np.sqrt(J_values))
    return Q_values - np.sqrt(J_values)

def V_LEPS_II_mod(rAB, x, debug = False):
    rAC = 3.742
    kC = 0.2025
    
    V_normal = V_LEPS_mod(rAB, rAC - rAB, debug)
    c = 1.154
    
    if debug:
        print("Harmonic : ", 2 * kC * (rAB - (rAC / 2 - x / c))**2)
    return V_normal + 2 * kC * (rAB - (rAC / 2 - x / c))**2
```

```python
min_1 = V_LEPS_II_mod(minima_1[0], minima_1[1], debug = False)
max_1 = V_LEPS_II_mod(saddle_point[0], saddle_point[1], debug = False)

print("min_1 : ", min_1)
print("max_1 : ", max_1)
```

```python
rAB_II_mod, x_mod = np.meshgrid(np.arange(0.2, 4.1, 0.05), np.arange(-2, 1.5, 0.05))
l = rAB_II_mod.flatten().shape[0]
LEPS_II_mod = np.zeros(l)
for i in np.arange(l):
    LEPS_II_mod[i] = V_LEPS_II_mod(rAB_II_mod.flatten()[i], x_mod.flatten()[i])
LEPS_II_mod = LEPS_II_mod.reshape(rAB_II.shape)
```

```python
T = 0.3
kB = 1
beta = 1 / (kB * T)
f_mod = lambda x, y : np.exp(-beta * V_LEPS_II_mod(x, y))
z = dblquad(f_mod, -np.inf, np.inf, lambda x:-np.inf, lambda x: np.inf)[0]
```

```python
rAB_II_mod, x_mod = np.meshgrid(np.arange(0.2, 4.1, 0.05), np.arange(-2, 1.5, 0.05))
l = rAB_II_mod.flatten().shape[0]
rho_LEPS_II_mod = np.zeros(l)
for i in np.arange(l):
    rho_LEPS_II_mod[i] = np.exp(-beta * V_LEPS_II_mod(rAB_II_mod.flatten()[i], x_mod.flatten()[i]))
rho_LEPS_II_mod = rho_LEPS_II_mod.reshape(rAB_II_mod.shape)
rho_LEPS_II_mod /= z
```
