```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
file_loc = '../../runs/langevin_2'
```

```python
vel = np.loadtxt(os.path.join(file_loc, "v.txt"))
steps = vel[:, 0]
vel = vel[:, 1]
```

```python
print(vel)
```

```python
p, be = np.histogram(vel, bins = 20, density = True)
coords = (be[1:] + be[:-1]) / 2
```

```python
m = 1
beta = 1 / 2
linsp = np.linspace(-6, 6, 1000)
expected_distribution = np.exp(- beta * m * linsp**2 / 2) / np.sqrt(2 * np.pi / beta)
# expected_distribution = np.exp(- beta * m * linsp**2 / 2) 
# expected_distribution /= expected_distribution.sum()
```

```python
fig = plt.figure(figsize = (7, 6))
plt.scatter(coords, p, color = 'purple', label = r'Obtained $\rho(v_i)$')
plt.plot(linsp, expected_distribution, color = 'orange', label = r'Expected $\rho(v_i)$')
plt.xlabel(r'$v_i$', fontsize = 15)
_ = plt.ylabel(r'$\rho(v_i)$', fontsize = 15)
plt.legend(loc = 'best')
_.set_rotation(0)
```
